```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(['science','notebook'])
```

```python
a = 1
b = 1
c = 1
```

```python
x = np.linspace(0,10,100)
y1 = a*x**2 + b*x + c
y2 = 0.5*x**2 + 0.5*x + c

x_i = np.linspace(0,10,10)
y_i = a*x_i**2 + b*x_i + c + np.random.randn(10)*10
```

```python
plt.plot(x,y1, label='Model $f(x)$')
plt.scatter(x_i,y_i, color='r', label='Data $(x_i, y_i)$')
plt.title('Model $f(x)=ax^2+bx+c$ for $a=1$, $b=1$, $c=1$', fontsize=12)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.savefig('plots/intro_image.png', dpi=200)
```
