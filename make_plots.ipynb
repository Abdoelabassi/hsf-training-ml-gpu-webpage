```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(['science','notebook'])
```

```python
a = 1
b = 1
c = 1

f = 1
g = 0.1 
```

```python
x = np.linspace(0,10,100)
y1 = a*x**2 + b*x + c
y2 = f*np.exp(0.5*x)

x_i = np.linspace(0,10,10)
y_i = a*x_i**2 + b*x_i + c + np.random.randn(10)*10

loss1 = np.sum((a*x_i**2 + b*x_i + c - y_i)**2)
loss2 = np.sum((f*np.exp(0.5*x_i) - y_i)**2)
```

```python
fig, ax = plt.subplots(1,2,figsize=(15,4))
ax[0].plot(x,y1, label='Model $f(x)$')
ax[0].scatter(x_i,y_i, color='r', label='Data $(x_i, y_i)$')
ax[0].set_title('Model $f(x)=ax^2+bx+c$ for $a=1$, $b=1$, $c=1$', fontsize=12)
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].legend()
ax[0].text(0.05,0.6,'Loss = {:.2f}'.format(loss1), transform=ax[0].transAxes, fontsize=14)

ax[1].plot(x,y2, label='Model $f(x)$')
ax[1].scatter(x_i,y_i, color='r', label='Data $(x_i, y_i)$')
ax[1].set_title('Model $f(x)=ae^{bx}$ for $a=1$, $b=0.5$', fontsize=12)
ax[1].set_xlabel('x')
ax[1].set_ylabel('y')
ax[1].legend()
ax[1].text(0.05,0.6,'Loss = {:.2f}'.format(loss2), transform=ax[1].transAxes, fontsize=14)
plt.savefig('plots/intro_image.png', dpi=200)
```
